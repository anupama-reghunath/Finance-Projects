# Loan Default Prediction – Probability of Default (PD) Model

In this notebook we build a simple Probability of Default (PD) model for a retail loan portfolio.

Goals:

1. Prepare features from the dataset for modelling(one-hot encoding for categorical variables).
2. Train a Logistic Regression model to estimate PD for each loan.
3. Evaluate model performance with confusion matrix, ROC/AUC and Default recall.
4. Experiment with different PD thresholds and see how it affects default recall vs accuracy.
5. Train a simple Random Forest model and compare with logistic regression.
6. Compute a simple Expected Loss using PD, LGD and EAD.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)

from sklearn.calibration import calibration_curve
```

```python
# Load preprocessed dataset
data = pd.read_csv("Preprocessed_Data.csv")

data.head()
```


Before modelling we want to understand the feature count, and verify if the data set is balanced for the target_variable(`loan_status`: 0 = non-default, 1 = default). 


```python
# Look at target distribution
data['loan_status'].value_counts(normalize=True).mul(100).round(2)
```

```python
data['loan_status'].value_counts().plot(kind='bar')
plt.title("Class counts: Non-default (0) vs Default (1)")
plt.xlabel("loan_status")
plt.ylabel("Count")
plt.show()
```

This is not unusual since one expects far more non-defaults than defaults. But we need to account for this imbalance in our model later.

```python
# Basic info
print(data.shape)
print(data.dtypes)

# Identify numeric and categorical columns
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = data.select_dtypes(include=["object"]).columns.tolist()

print("\nNumber of numeric features:", len(numeric_cols))
print("Number of categorical features:", len(categorical_cols))
```

```python
# Quick cardinality check
categorical_cardinality = data[categorical_cols].nunique().sort_values(ascending=False)
display(categorical_cardinality.to_frame("n_unique_categories"))
```

We apply one-hot encoding (dummy variables) to just the categorical columns and re-combine them into a single modelling dataset `data_prep`.

```python
# Separate numeric and categorical columns
data_num = data.select_dtypes(exclude=['object'])
data_cat = data.select_dtypes(include=['object'])
```

```python
# One-hot encode categorical vars
data_cat_dummies = pd.get_dummies(data_cat, drop_first=True) # drop_first to avoid dummy variable trap for grade = {A, B, C}, if grade_B(0) and grade_C(0) are included, grade_A is implied(1).

# Combine back
data_prep = pd.concat([data_num, data_cat_dummies], axis=1)

data_prep.head()
```

```python
data.shape,data_prep.shape
```

All looks good. Now we split into training and test sets (60% train, 40% test).

```python
X = data_prep.drop(columns=['loan_status']) # Features
y = data_prep['loan_status'] # Target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42, stratify=y
)

X_train.shape, X_test.shape
```

First lets fit a Logistic Regression model, which directly outputs a probability between 0 and 1. This is a natural choice for a PD model.


```python
log_clf = LogisticRegression(
    solver='lbfgs',
    max_iter=500 # to avoid convergence warnings.
)

log_clf.fit(X_train, y_train)
```

```python
coef_df = pd.DataFrame({
    "feature": X_train.columns,
    "coefficient": log_clf.coef_[0]
}).sort_values("coefficient", ascending=False)

display(coef_df.head(15))
```

```python
# Predicted probabilities for the positive class (default = 1)
y_proba_log = log_clf.predict_proba(X_test)[:, 1]

# Default threshold 0.5
y_pred_log_50 = (y_proba_log >= 0.5).astype(int)

print("Accuracy (threshold 0.5):", accuracy_score(y_test, y_pred_log_50))
print("\nClassification report (threshold 0.5):")
print(classification_report(y_test, y_pred_log_50, target_names=["Non-default", "Default"]))

print("Confusion matrix (threshold 0.5):")
print(confusion_matrix(y_test, y_pred_log_50))
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba_log)
auc_log = roc_auc_score(y_test, y_proba_log)

plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {auc_log:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--", label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve – Logistic Regression")
plt.legend()
plt.show()
```


For credit risk, threshold choice matters.
Higher threshold means fewer loans classified as default=> higher accuracy on non-defaults, but lower default recall.

Lower threshold=> we catch more defaults (higher default recall) but more false positives.


```python

thresholds = np.linspace(0.1, 0.9, 9)

results = []

for thr in thresholds:
    y_pred_thr = (y_proba_log >= thr).astype(int)
    
    acc = accuracy_score(y_test, y_pred_thr)
    
    # classification_report with output_dict=True to get per-class recall
    rep = classification_report(
        y_test, y_pred_thr,
        target_names=["Non-default", "Default"],
        output_dict=True
    )
    
    recall_nondefault = rep["Non-default"]["recall"]
    recall_default = rep["Default"]["recall"]
    
    results.append((thr, acc, recall_nondefault, recall_default))

results_df = pd.DataFrame(
    results,
    columns=["threshold", "accuracy", "recall_nondefault", "recall_default"]
)

results_df
```

```python
plt.plot(results_df["threshold"], results_df["recall_default"], label="Recall (defaults)")
plt.plot(results_df["threshold"], results_df["recall_nondefault"], label="Recall (non-defaults)")
plt.plot(results_df["threshold"], results_df["accuracy"], label="Accuracy")

plt.xlabel("Probability Threshold")
plt.ylabel("Score")
plt.title("Effect of Threshold on Recall and Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

We can use **RandomForestClassifier** from scikit-learn as a simple tree-ensemble model.

We compare its ROC–AUC with logistic regression.


```python
rf_clf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf_clf.fit(X_train, y_train)

# Probabilities and AUC
y_proba_rf = rf_clf.predict_proba(X_test)[:, 1]
auc_rf = roc_auc_score(y_test, y_proba_rf)

print("Logistic Regression AUC:", auc_log)
print("Random Forest AUC:", auc_rf)
```

```python
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)

plt.plot(fpr, tpr, label=f"Logistic (AUC = {auc_log:.3f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {auc_rf:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--", label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curves – Logistic vs Random Forest")
plt.legend()
plt.grid(True)
plt.show()
```

For each loan a Simple Expected Loss (EL) is defined as:

$$
EL = PD \times LGD \times EAD
$$

Here,
- PD = predicted probability of default from our model  
- LGD = assumed loss given default (e.g. 20%)  
- EAD = exposure at default, here we approximate with `loan_amnt`

```python
# Use logistic regression PDs for simplicity
pd_series = pd.Series(y_proba_log, index=y_test.index, name="PD")

# Grab loan amount from original data (align indices with y_test)
loan_amnt_test = data.loc[y_test.index, "loan_amnt"]

LGD = 0.2  # assume 20% loss given default

el = pd_series * LGD * loan_amnt_test
total_el = el.sum()

print(f"Total Expected Loss (Logistic PD, LGD={LGD:.2f}): {total_el:,.2f}")
```

```python
pd_series_rf = pd.Series(y_proba_rf, index=y_test.index, name="PD_rf")
el_rf = pd_series_rf * LGD * loan_amnt_test
print(f"Total Expected Loss (Random Forest PD, LGD={LGD:.2f}): {el_rf.sum():,.2f}")
```
